```python
import pandas as pd  
import numpy as np 
```

```python
data=pd.read_csv('E:\\FinalProject\\notebooks\\data\\cubic_zirconia.csv')
```

```python
data.drop(data.columns[0],axis=1,inplace=True)
```

```python
data.head()
```

```python
X=data.drop(labels=["price"],axis=1)
```

```python
X
```

```python
y=data[["price"]]
```

```python
cate_col=X.select_dtypes(include="object")
```

```python
Num_col=X.select_dtypes(exclude="object")
```

```python
cut_categories=['Fair', 'Good', 'Very Good','Premium','Ideal']
color_categories=['D', 'E', 'F', 'G', 'H', 'I', 'J']
clarity_categories=['I1','SI2','SI1','VS2','VS1','VVS2','VVS1','IF']
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OrdinalEncoder
from sklearn.pipeline import Pipeline 
from sklearn.compose import ColumnTransformer
```

```python
num_pipeline=Pipeline(
        
        
    steps=[
        
        
        ('imputer',SimpleImputer()),
        ('Scaler',StandardScaler())
    ]
           
        
    
    
)
```

```python
cate_pipeline=Pipeline(
    
    steps=[
        ('imputer',SimpleImputer(strategy='most_frequent')),
        ('OrdinalEncoder',OrdinalEncoder(categories=[cut_categories,color_categories,]))
        
        
        
    ]
)
```

```python
Preprocessor=ColumnTransformer(
    [
     
     ('num_pipeline',num_pipeline,Num_col),
     ('cate_pipeline',cate_pipeline,cate_col)
    
    
    
    
    ]
)


```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=30)
```

```python
Preprocessor.fit_transform(X_train)
```

```python
Preprocessor.transform(X_test)
```
